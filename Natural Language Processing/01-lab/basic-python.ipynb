# Question 1. Create a function to find Min, Max, Mean, and Median on a list.

Input(parameters):

1. name of the function(String) and
2. List of values

Output:

1. Metric value

PS. sort vs sorted [link text](https://www.geeksforgeeks.org/python-difference-between-sorted-and-sort/)


```python
vals = [7, 14, 81, 56, 9, 17, 43]


def getMetric(metric, ListVal):
    if metric == "mean":
        return sum(ListVal) / len(ListVal)
    elif metric == "median":
        ListVal.sort()
        if len(ListVal) % 2 == 0:
            return (
                ListVal[int(len(ListVal) / 2)] + ListVal[int(len(ListVal) / 2) - 1]
            ) / 2
        else:
            return ListVal[int(len(ListVal) / 2)]

    return None


mean = getMetric("mean", vals)
med = getMetric("median", vals)
print(mean)
print(med)
```

# Question 2. Count the number of words appearing in text.

- read text file from sampleText.txt
- Count number of each word by the appearance in text
- the result should show like below

Result

1. sed 5650
2. in 4748
3. amet 4594
4. sit 4541
5. id 3853
6. ...

PS. To sort Dict: [article](https://thispointer.com/sort-a-dictionary-by-value-in-python-in-descending-ascending-order/)


```python
import re

# inPath = "/content/sampleText.txt" # for Google Colab
inPath = "sampleText.txt"  # for local

with open(inPath) as f:  # read as one long string
    contents = f.read()

words = contents.split()
for w in words:
    w = re.sub(r"[^A-Za-z]", "", w)

print(words[:10])
```

```python
counts = {}

for w in words:
    if w in counts:
        counts[w] += 1
    else:
        counts[w] = 1

print(sorted(counts.items(), key=lambda x: x[1], reverse=True))

for k, v in sorted(counts.items(), key=lambda x: x[1], reverse=True):
    print(k, v)
```

```python
for k in sorted(counts, key=counts.get, reverse=True):
    print(k, counts[k])
```
